# View Vs Copy ==> When we slice a sub-array from an array , it may be done by two ways . 

```python
import numpy as np 
a = np.array([10 , 20 , 30 , 40 , 50 , 60 , 70 , 80 ])
b = a[3:6] 
b[:] = 0 
a
```

# copy 

```python
a = np.array([10 , 20 , 30 , 40 , 50 , 60 , 70 , 80 ])
b = a[3:6].copy() 
b[:] = 0 
a
```

```python
b
```
