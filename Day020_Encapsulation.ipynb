```python
class BankAccount():
    def __init__(self,accountHolder,balance):
        self.__accountHolder=accountHolder
        self.__balance=balance
    
    def getAccountHolder(self):
        return self.__accountHolder
    
    def getBalance(self):
        return self.__balance
        
    def deposite(self,amount):
        if amount > 0:
            self.__balance+=amount
            return f'deposite successful, the new balance is {self.__balance}'
        else:
            return 'Invalid amount'
    
    def withdraw(self,amount):
        if amount<=self.__balance:
            self.__balance-=amount
            return f'withdraw successful, the new balance is {self.__balance}'
        else:
            return 'Insufficeint Balance'

account1=BankAccount('Ash',5000)
print(account1.getAccountHolder())
print(account1.getBalance())


print(account1.deposite(1000))
print(account1.withdraw(500))
print(account1.deposite(1000))
print(account1.getBalance())
print(account1.getAccountHolder())
```

```python
class Student():
    def __init__(self,name,grade):
        self.name=name
        self.__grade=grade
    
    def getGrade(self):
        return self.__grade
    
        
    def setGrade(self,grade):
        if grade >= 0 and grade <= 100:
            self.__grade=grade
            return f'update successful, the new grade is {self.__grade}'
        else:
            return 'Invalid grade.the grade 0 to 100'


std=Student('Ash',5)
print(std.name)
print(std.getGrade())


print(std.setGrade(100))
print(std.setGrade(500))

print(std.name)
print(std.getGrade())
```

```python
from datetime import datetime
import time
class patientRecord():
    def __init__(self,name,age,medicalHistory=''):
        self.__name=name
        self.__age=age
        self.medicalHistory=medicalHistory+f',"{datetime.now()}"\n'
    
    def get_details(self):
        return self.__name,self.__age,self.medicalHistory
    
    
    def add_medicalHistory(self,medicalHistory):
        if medicalHistory:
            self.medicalHistory+=medicalHistory+f',"{datetime.now()}"\n'
            return f'update successful, the new medical history is \n{self.medicalHistory}'
        else:
            return f'Nothing is updated in Medical History'
    
p=patientRecord('Ash',18,'Fever')
PatientDetails=p.get_details()
print(f'Patient Records: \n\tName -> {PatientDetails[0]} \n\tAge -> {PatientDetails[1]}')
time.sleep(10)
print(p.add_medicalHistory('Maleria'))
PatientDetails=p.get_details()
print(f'Patient Records: \n\tName -> {PatientDetails[0]} \n\tAge -> {PatientDetails[1]}\n\tMedical History -> {PatientDetails[2]}')
```
