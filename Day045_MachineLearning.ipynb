Linear Regression

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['sex']=lb.fit_transform(df['sex'])
df['smoker']=lb.fit_transform(df['smoker'])
df['region']=lb.fit_transform(df['region'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns=['charges'])
y=df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pridt=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pridt)
```

```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('Day028_Sample_data_covid_toy.csv')
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer()
df['fever']=si.fit_transform(df[['fever']])
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_preidict=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_preidict)
```

# New DATA

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df=pd.read_csv('Day028_placement.csv')
print(df.head())

print(df.isna().sum())

x=df.drop(columns=['placed'])
y=df['placed']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()

print(lr.fit(x_train,y_train))

y_pridict=lr.predict(x_test)

from sklearn.metrics import accuracy_score

print('accuracy percentage --->',accuracy_score(y_test,y_pridict))
```
